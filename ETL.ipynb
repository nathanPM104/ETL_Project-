```python
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

```python
#load in suicide rates data and create pandas datafile
suicide_file = "Resources/SuicideRates.csv"
suicide_df = pd.read_csv(suicide_file)
suicide_df.head(50)
```

```python
suicide_df_2016 = suicide_df.loc[suicide_df['year']==2016]
suicide_df_2016 = suicide_df_2016.drop(['country-year','sex'], axis=1)
suicide_df_2016 = suicide_df_2016.groupby(['country'])
suicide_df_2016 = suicide_df_2016['suicides_no'].sum()
suicide_df_2016 = pd.DataFrame(suicide_df_2016).reset_index()
suicide_df_2016.columns = ['country', 'suicidecount']
```

```python
suicide_df_2016.head()
```

```python
#load in coffee data and create pandas datafile
coffee_file = "Resources/domestic-consumption.csv"
coffee_df = pd.read_csv(coffee_file)
coffee_df.head(100)
```

```python
coffee_df_2016 = coffee_df[['domestic_consumption', '2016']].copy()
coffee_df_2016.head()
```

```python
coffee_df_2016.rename(columns={"domestic_consumption" : "country", "2016" : "coffee_consumption"}, inplace=True)
coffee_df_2016.head()
```

```python
#load in happy and beers data and create pandas datafile
HappinessAlcohol_file = "Resources/HappinessAlcoholConsumption.csv"
HappinessAlcohol_df = pd.read_csv(HappinessAlcohol_file)
HappinessAlcohol_df.head()
```

```python
HappinessAlcohol_df = HappinessAlcohol_df.drop(['Region','Hemisphere'],axis=1)
```

```python
HappinessAlcohol_df = HappinessAlcohol_df.rename(columns={'Country':'country','HappinessScore': "happinessscore",
                                                'HDI':'hdi','GDP_PerCapita':'gdp_percapita','Beer_PerCapita':'beer_percapita',
                                                'Spirit_PerCapita':'spirit_percapita','Wine_PerCapita':'wine_percapita'})
```

```python
HappinessAlcohol_df.head()
```

```python
conn = f'postgresql://postgres:{password}@localhost:5432/ETL'
```

```python
engine = create_engine(conn, echo=True)
sqlite_connection = engine.connect()
```

```python
coffee_df_2016.to_sql('coffee_df_2016', con=engine, if_exists='append', index=False)

suicide_df_2016.to_sql('suicide_df_2016', con=engine, if_exists='append', index=False)

HappinessAlcohol_df.to_sql('happinessalcohol_df', con=engine, if_exists='append', index=False)
```
